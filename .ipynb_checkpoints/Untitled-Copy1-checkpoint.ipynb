```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
colnames = ['classes', 'cap-shape', 'cap-surface', 'cap-color','bruises', 'odor', 'gill-attachment', 'gill-spacing','gill-size', 'gill-color', 'stalk-shape', 'stalk-root','stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring', 'stalk-color-below-ring','veil-type', 'veil-color', 'ring-number', 'ring-type', 'spore-print-color', 'population', 'habitat']
shrooms = pd.read_csv('agaricus-lepiota.data', names=colnames, header=None)
```

```python
shrooms.head(2)
```

```python
shrooms['classes'] = shrooms['classes'].replace({'p': 'poisonous', 'e': 'edible'})
```

```python
shrooms['classes'] = shrooms['classes'].replace({'p': 'poisonous', 'e': 'edible'})
shrooms['cap-shape'] = shrooms['cap-shape'].replace({'b': 'bell', 'c': 'conical','x': 'convex', 'f': 'flat','k': 'knobbed', 's': 'sunken'})
shrooms['cap-surface'] = shrooms['cap-surface'].replace({'f': 'fibrous', 'g': 'grooves','y': 'scaly', 's': 'smooth'})
shrooms['cap-color'] = shrooms['cap-color'].replace({'n': 'brown', 'b': 'buff','c': 'cinnamon', 'g': 'gray','r': 'green', 'p': 'pink', 'u':'purple', 'r':'red', 'w':'white', 'y':'yellow'})
shrooms['bruises'] = shrooms['bruises'].replace({'t': 'bruises', 'f': 'no'})
shrooms['odor'] = shrooms['odor'].replace({'a': 'almond', 'l': 'anise','c': 'creosote', 'y': 'fishy','f': 'foul', 'm': 'musty', 'n':'none', 'p': 'pungent', 's':'spicy'})
shrooms['gill-attachment'] = shrooms['gill-attachment'].replace({'a': 'attached', 'd': 'descending','f': 'free', 'n': 'notched'})
shrooms['gill-spacing'] = shrooms['gill-spacing'].replace({'c': 'close', 'w': 'crowded','d': 'distant'})
shrooms['gill-size'] = shrooms['gill-size'].replace({'b': 'broad', 'n': 'narrow'})
shrooms['gill-color'] = shrooms['gill-color'].replace({'k': 'black', 'n': 'brown','b': 'buff', 'h': 'chocolate','g': 'gray', 'r': 'green', 'o':'orange', 'p':'pink', 'u':'purple', 'r':'red', 'w':'white', 'y':'yellow'})
shrooms['stalk-shape'] = shrooms['stalk-shape'].replace({'e': 'enlarging', 't': 'tapering'})
shrooms['stalk-root'] = shrooms['stalk-root'].replace({'b': 'bulbous', 'c': 'club','u': 'cup', 'e': 'equal','z': 'rhizomorphs', 'r': 'rooted', '?':'missing'})
shrooms['stalk-surface-above-ring'] = shrooms['stalk-surface-above-ring'].replace({'f': 'fibrous', 'y': 'scaly','k': 'silky', 's': 'smooth'})
shrooms['stalk-surface-below-ring'] = shrooms['stalk-surface-below-ring'].replace({'f': 'fibrous', 'y': 'scaly','k': 'silky', 's': 'smooth'})
shrooms['stalk-color-above-ring'] = shrooms['stalk-color-above-ring'].replace({'n': 'brown', 'b': 'buff','c': 'cinnamon', 'g': 'gray','o': 'orange', 'p': 'pink', 'r':'red', 'w':'white', 'y':'yellow'})
shrooms['stalk-color-below-ring'] = shrooms['stalk-color-below-ring'].replace({'n': 'brown', 'b': 'buff','c': 'cinnamon', 'g': 'gray','o': 'orange', 'p': 'pink', 'r':'red', 'w':'white', 'y':'yellow'})
shrooms['veil-type'] = shrooms['veil-type'].replace({'p': 'partial', 'u': 'universal'})
shrooms['veil-color'] = shrooms['veil-color'].replace({'n': 'brown', 'o': 'orange','w': 'white', 'y': 'yellow'})
shrooms['ring-number'] = shrooms['ring-number'].replace({'n': 'none', 'o': 'one','t': 'two'})
shrooms['ring-type'] = shrooms['ring-type'].replace({'c': 'cobwebby', 'e': 'evanescent','f': 'flaring', 'l': 'large','n': 'none', 'p': 'pendant', 's':'sheathing', 'z':'zone'})
shrooms['spore-print-color'] = shrooms['spore-print-color'].replace({'k': 'black', 'n': 'brown','b': 'buff', 'h': 'chocolate','r': 'green', 'o': 'orange', 'u':'purple', 'w':'white', 'y':'yellow'})
shrooms['population'] = shrooms['population'].replace({'a': 'abundant', 'c': 'clustered','n': 'numerous', 's': 'scattered','v': 'several', 'y': 'solitary'})
shrooms['habitat'] = shrooms['habitat'].replace({'u': 'urban', 'g': 'grasses','m': 'meadows', 'l': 'leaves','w': 'waste', 'd': 'woods', 'p':'paths'})
```

```python
# Import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
# Import train_test_split
from sklearn.model_selection import train_test_split
# Import accuracy_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.tree import export_graphviz
from sklearn.metrics import roc_auc_score
```

```python
#subset data into dependent and independent variables x,y
LABELS = ['classes_edible', 'classes_poisonous']
FEATURES = [a  for a in dummy_shrooms.columns if a not in LABELS ]
y = dummy_shrooms[LABELS[0]]
X= dummy_shrooms[FEATURES]

# Split dataset into 80% train, 20% test
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, stratify=y, random_state=1)
```

```python
#X = shrooms.drop(['classes'], axis=1)
X=shrooms.drop(['classes'], axis=1)
y=shrooms['classes']

# Split dataset into 80% train, 20% test
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, stratify=y, random_state=1)
```

```python
# Instantiate 
tree_shrooms = DecisionTreeClassifier()

# Fit dt to the training set
tree_shrooms.fit(X_train,y_train)

# Predict test set labels
y_pred = tree_shrooms.predict(X_test)
```

```python
predTree = tree_shrooms.predict(X_test)

print (f'Predicted results\n{predTree [0:5]}\n')
print (f'Actual results\n{y_test [0:5]}')
```
